```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_file = "Resources/schools_complete.csv"
student_file = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_file)
student_data = pd.read_csv(student_file)

# Combine the data into a single dataset
school_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Rename the columns so they're nicer
school_df_rename = school_df.rename(columns={'student_name':'Student Name',
                                      'gender':'Gender', 
                                      'grade':'Grade', 
                                      'school_name':'School Name',
                                      'reading_score':'Reading Score', 
                                      'math_score':'Math Score', 
                                      'type':'School Type', 
                                      'size':'School Size',
                                      'budget':'School Budget'
                                      })
school_df_rename.head(1)
```

```python
#Reset the dataframe index to Student ID and get a simpler dataframe variable name
df = school_df_rename.set_index('Student ID')
df.head(1)
```

```python
#Add some bins for pass/fail in math, reading and overall for later analysis. 
#We could do this with loc, but it's uglier.
# bins = [0, 69, 100]
# labels = [0, 1]
# df['% Passing Reading'] = pd.cut(df["Reading Score"], bins, labels=labels)
# df['% Passing Math'] = pd.cut(df["Math Score"], bins, labels=labels)

# grade_avg = (df['Math Score'] + df['Reading Score'])/2
# df['Overall Passing Rate'] = pd.cut(grade_avg, bins, labels=labels)
df['Overall Average Score'] = (df['Reading Score'] + df['Math Score'])/2
df.head()
```

```python
#Calculate the number of schools
num_schools = len(df['School Name'].unique())
```

```python
#Calculate the total number of students
num_students = len(df)
```

```python
#Calculate the total budget
budget_total = df['School Budget'].unique().sum() #We could just pull these from the school_data df too
```

```python
#Calculate the average math score
#Calculate the average reading score
math_mean = df['Math Score'].mean()
reading_mean = df['Reading Score'].mean()
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate = df['Overall Average Score'].mean()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_math = df.loc[df['Math Score'] >= 70, :]
pass_read = df.loc[df['Reading Score'] >= 70, :]
# pass_math = df.loc[df['Math']=='1' # alternative method using bins
# pass_read = (df.loc[df['Reading']=='1', :]) # alternative method using bins
pct_pass_math = len(pass_math)*100/num_students
pct_pass_read = len(pass_read)*100/num_students
```

```python
# Create a dataframe to hold the above results
#Format the percentages in one go

district_summary = pd.DataFrame({"Number of Schools":[num_schools],
                               "Number of Students":[num_students],
                               "Total Budget":[budget_total],
                               "Avgerage Math Score":[math_mean],
                               "Avgerage Reading Score":[reading_mean],
                               "% Passing Math":[pct_pass_math],
                               "% Passing Reading":[pct_pass_read],
                               "Overall Passing Rate": [passing_rate]})
```

```python
# Optional: give the displayed data cleaner formatting
district_summary['Number of Students'] = district_summary['Number of Students'].map('{:,}'.format)
district_summary['Total Budget'] = district_summary['Total Budget'].map('${:,}'.format)
district_summary['Avgerage Math Score'] = district_summary['Avgerage Math Score'].map('{:.2f}'.format)
district_summary['Avgerage Reading Score'] = district_summary['Avgerage Reading Score'].map('{:.2f}'.format)
district_summary['% Passing Math'] = district_summary['% Passing Math'].map('{:.2f}%'.format)
district_summary['% Passing Reading'] = district_summary['% Passing Reading'].map('{:.2f}%'.format)
district_summary['Overall Passing Rate'] = district_summary['Overall Passing Rate'].map('{:.2f}%'.format)
district_summary
```
