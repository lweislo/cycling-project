```python
# # np.savetxt('women2019_2.txt', womenfn.values)
# filename = "women_2.txt"

def macroout(filename, dataset):
    for index, row in dataset.iteritems():
        with open(filename, "a+") as f:
            f.write(str(row) + '\n')

```

```python
macroout('Women2019LN.txt',womenfn)
macroout('WTmen2019FN.txt',menwtfn)
macroout('PCmen2019FN.txt',menpcfn)
macroout('Women2019LN.txt',womenln)
macroout('WTmen2019LN.txt',menwtln)
macroout('PCmen2019LN.txt',menpcln)
```
