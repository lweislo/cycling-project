```python
# Multiple last names from certain countries often only go by the first surname except de la ... 
for index, row in df3.iterrows():
    if len(row['len_last']) > 1 and row['country_short'] in ['(Spa)','(Col)','(Por)','(Mex)'] and row['len_last'][0]!='De':
        
        row['lnwild'] = row['len_last'][0]
        df3.loc[index, 'lnwild'] = row['len_last'][0]
        print(df3.loc[index, 'lnwild'])
# df3.loc[df3['country_short_y'] == '(Spa)', ['lnwild']] = (df3['len_last'] + str('*')).where((df3['len_last']).str.len() == 1)
        
```

```python
# Danes tend to keep both surnames but not always
for index, row in df3.iterrows():
    if len(row['len_last']) > 1 and row['country_short'] == '(Den)':
        df3.loc[index, 'lnwild'] = df3.loc[index, 'last_name']
        print(df3.loc[index, 'lnwild'])
```

```python
# For other countries, need to make sure we don't separate names
for index, row in df3.iterrows():
    if len(row['len_last']) > 1 and row['country_short'] not in ['(Spa)','(Col)','(Por)', '(Mex)'] and row['len_last'][0] in ['De','Mc','Van', 'Ten', 'van', 'Le', 'El', 'de','Paret']:
        df3.loc[index, 'lnwild'] = df3.loc[index, 'last_name']
        print(df3.loc[index, 'lnwild'])
```

```python
# Simple last names stay the same
for index, row in df3.iterrows():
    if len(row['len_last']) == 1:
        df3.loc[index, 'lnwild'] = df3.loc[index, 'last_name']
        print(df3.loc[index, 'lnwild'])
```

```python
# The de la's
for index, row in df3.iterrows():
    if len(row['len_last'][0:3]) >= 3 and row['len_last'][0]=='De':
        ln = row['len_last'][0:3]
        last = ' '.join(map(str, ln))
        df3.loc[index, 'lnwild'] = last
        print(df3.loc[index, 'lnwild'])
```

```python
# # np.savetxt('women2019_2.txt', womenfn.values)
# filename = "women_2.txt"

def macroout(filename, dataset):
    for index, row in dataset.iteritems():
        with open(filename, "w+") as f:
            f.write(str(row) + '\n')

```

```python
macroout('Women2019FN.csv',womenfn)
macroout('WTmen2019FN.csv',menwtfn)
macroout('PCmen2019FN.csv',menpcfn)
macroout('Women2019LN.csv',womenln)
macroout('WTmen2019LN.csv',menwtln)
macroout('PCmen2019LN.csv',menpcln)
```
