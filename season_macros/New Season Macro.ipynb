```python
# Multiple last names from certain countries often only go by the first surname except de la ... 
for index, row in df3.iterrows():
    if len(row['len_last']) > 1 and row['Country'] in double_surnames and row['len_last'][0]!='De':
        
        row['lnwild'] = row['len_last'][0]
        df3.loc[index, 'lnwild'] = row['len_last'][0]
        print(df3.loc[index, 'lnwild'])
# df3.loc[df3['country_short_y'] == '(Spa)', ['lnwild']] = (df3['len_last'] + str('*')).where((df3['len_last']).str.len() == 1)
        
```

```python
# Danes tend to keep both surnames but not always
dbl = ['(Den)','(Swe)','(Nor)','(Fin)']


for index, row in df3.iterrows():
    for item in dbl:
        if len(row['len_last']) > 1 and row['Country'] == item:
            df3.loc[index, 'lnwild'] = df3.loc[index, 'Last Name']
            print(df3.loc[index, 'lnwild'])
```

```python
# For other countries, need to make sure we don't separate names
for index, row in df3.iterrows():
    if len(row['len_last']) > 1 and row['Country'] not in double_surnames and row['len_last'][0] in ['De','Mc','Van', 'Ten', 'van', 'Le', 'El', 'de','Paret','Longo','Ferrand','Geoghegan','ten','Janse','Mirza','Abraham','Cordon','Uiterwijk','Du']:
        df3.loc[index, 'lnwild'] = df3.loc[index, 'Last Name']
        print(df3.loc[index, 'lnwild'])
```

```python
# Simple last names stay the same
for index, row in df3.iterrows():
    if len(row['len_last']) == 1:
        df3.loc[index, 'lnwild'] = df3.loc[index, 'Last Name']
        print(df3.loc[index, 'lnwild'])
```

```python
# The de la's
for index, row in df3.iterrows():
    if len(row['len_last'][0:3]) >= 3 and row['len_last'][0]=='De' or row['len_last'][0]=='Da':
        ln = row['len_last'][0:3]
        last = ' '.join(map(str, ln))
        df3.loc[index, 'lnwild'] = last
        print(df3.loc[index, 'lnwild'])
```

```python
# This creates the UCI Code/Rider Name find/replace macros
ucidf = pd.DataFrame()
ucidf['cat'] = df3['Category']
ucidf['ln'] = df3['Last Name']
ucidf['tn'] = df3['Team Name']
ucidf['code'] = df3['UCIID'].astype(str)
ucidf['name'] = df3['cn_name']
ucidf['leading'] = 'Selection.Replace What:="'
ucidf['middle'] = '", Replacement:="'
ucidf['ending'] = '", LookAt:=xlWhole, SearchOrder:=xlByRows, MatchCase:=False'

ucidf['macro'] = ucidf['leading'] + ucidf['code'] + ucidf['middle'] + ucidf['name'] + ucidf['ending']
ucidf.sort_values(['cat','tn','ln'])
ucidf
```

```python
# # np.savetxt('women2019_2.txt', womenfn.values)
# filename = "women_2.txt"

def macroout(filename, dataset):
    for index, row in dataset.iteritems():
        with open(filename, "a") as f:
            f.write(str(row) + '\n')

```
